```python
# change column names for easier data handling

df2 = df.rename(columns={"What role(s) are you volunteering for?": "Roles", "What experience do you have with the roles you're volunteering for?": "Experience"})

df2.tail()
```

```python
# filter Roles column based on number of words
# df4: > 20
# df3: < 20

df3 = df2[df2['Roles_num'] < 20]

df3
```

```python
# df3a created in parallel to df3
# will specifically focus on roles column

df3a = df2[df2['Roles_num'] < 20]

df3a.tail()
```

```python
role = df3a['Roles']

role.tail()
```

```python
role.head()
```

```python
# loop through column, split "roles" into separate lists
# note: results in nested list

split_role = []

for cell in role:
    split_role.append(cell.split(","))
    
print(split_role)
```

```python
# function to flatten list

def flatten(lst):
    for i in lst:
        if isinstance(i, (list, tuple)):
            for ii in flatten(i):
                yield ii
        else:
            yield i
            
new_list = list(flatten(split_role))
print(new_list)
```

```python
creative_role_list = []

for string in new_list:
    if 'reati' in string:
        creative_role_list.append(string)

len(creative_role_list)
```

```python
# convert new_list into data frame

role_df = pd.DataFrame(new_list)

role_df.tail()
```
